# Extraindo informações estatícas sobre as turmas
Nesse seção iremos extrair informações estatiscas sobre as turmas, assim, teremos dados sobre componentes curriculasres, cursos, docentes, avaliação, etc.

```python
import pandas as pd
import numpy as np
```

```python
cursos = pd.read_csv("data/cursos-de-graduacao.csv", sep=';')
```

```python
id_cet = 10320810

cursos[cursos.nome == 'CIÊNCIAS E TECNOLOGIA']
```

```python
# importando estrutura curriculas
estrutura = pd.read_csv("data/estruturas-curriculares.csv", sep=";")
```

```python
estrutura[estrutura.id_curso == id_cet]
```

```python
curriculo = pd.read_csv("data/curriculo-componente-graduacao.csv", sep=";")
```

```python
curriculo_cet = curriculo[curriculo.id_curriculo == 115802434]
```

```python
# importando dados do componente curriculas
disciplinas = pd.read_csv("data/componentes-curriculares-presenciais-ok.csv")
disciplinas.iloc[:, ]
```

```python
disciplinas_cet = disciplinas[disciplinas.id_componente.isin(curriculo_cet.id_componente_curricular)]
```

```python
cols = ['id_componente', 'tipo_componente', 'codigo', 'nome', 'unidade_responsavel', 'ementa', \
       'bibliografia', 'objetivos', 'conteudo']

disciplinas_cet = disciplinas_cet[cols]
```

```python
disciplinas_cet.to_csv('outputs/displicinas_cet.csv', sep=';', index_label=False)
```
