# Mesclando os dados de Matrículas, Turmas e Disciplinas
## Objetivo
Esse notebook visa pegar informações que relacionam turmas, matrículas e disciplinas para facilitar 
a visualização dos dados.

```python
import pandas as pd
```

```python
matr = pd.read_csv("data/outputs/matriculas_cet_20172.csv", sep=';')
disciplinas = pd.read_csv("data/componentes-curriculares-presenciais-ok.csv", index_col=0)
turmas = pd.read_csv("data/turmas-20172.csv", sep=";")
```

```python
matr.head()
```

```python
disciplinas.head()
```

```python
turmas.head(2)
```

## Pegando só as colunas desejadas

```python
cols = ['id_turma', 'id_componente_curricular', 'ano', 'periodo']
turmas = turmas[cols]
```

```python
cols = ['id_componente', 'codigo', 'nome', 'curso_componente']
disciplinas = disciplinas[cols]
disciplinas = disciplinas.rename(index=str, columns={'id_componente': 'id_componente_curricular'})
```

## Realizando merge de Matriculas e Turmas

```python
matr.shape
```

```python
matriculas = matr.merge(turmas, how='left', left_on='id_turma', right_on='id_turma')
matriculas.shape
```

```python
matriculas = matriculas.drop_duplicates()
matriculas.shape
```

## Realizando merge de Matriculas e Disciplinas

```python
matr_ect = matriculas.merge(disciplinas, how='left', left_on='id_componente_curricular', right_on='id_componente', )
matr_ect.shape
```
