# Montando os dados da Turma
Nesse notebooks iremos juntas os dados das turmas e tentar montar essa tabela

```python
import pandas as pd
import numpy as np
```

```python
turmas = pd.read_csv('data/turmas-2017.1.csv', sep=';')
turmas.head(100)
```

```python
turmas.columns
```

```python
# importando disciplinas
disciplinas = pd.read_csv("outputs/displicinas_cet.csv", sep=';')
```

```python
turmas_base = turmas[['id_turma', 'id_docente_interno', 'id_componente_curricular']]
turmas_base.shape
```

```python
disc_base = ['id_componente', 'nome', 'codigo']
disc_ect = disciplinas[disc_base]
disc_ect.shape
```

```python
pd.merge(left=turmas_base, right=disc_ect, left_on='id_componente_curricular', right_on='id_componente', how='inner').shape
```

```python
# Lendo turmas de CeT
matr_cet = pd.read_csv("outputs/turmas_dados_2017_2.csv", sep=";")
matr_cet.shape
```

```python
turma_ax = pd.merge(left=turmas_base, right=matr_cet, how='inner')
turma_ax.head()
```

```python
turmas_com = pd.merge(left=turma_ax, right=disc_ect, left_on='id_componente_curricular', right_on='id_componente', how='inner')
turmas_com[['id_turma', 'nome', 'id_docente_interno', 'qnt_disc']].head()
```

```python
# importando professores
prof = pd.read_csv("data/docentes.csv", sep=';')
prof.head(2)
```

```python
turma_simp = turmas_com[['id_turma', 'nome', 'id_docente_interno', 'qnt_disc']]
turma_simp['id_docente'] = turma_simp[['id_docente_interno']].applymap(np.int64)
```

```python
pd.merge(left=turma_simp, right=prof, left_on='id_docente', right_on='')
```
